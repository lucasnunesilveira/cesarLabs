# Task 
### Calcule a quantidade de avaliações por marca
### Calcule a media das avalições do produto a vista( 10% do valor Original)
### Transforme as palavas das colunas Style para português
### conte quantos pratos (variedades) possuem Noodles

```python
import pandas as pd 
dados_do_lamen = pd.read_csv("ramen-ratings.csv")
dados_do_lamen.head(10)
```

```python
dados_do_lamen.columns = ['reveja','marca','variedade','estilo','pais','notas','top_10']
dados_do_lamen
```

```python
#Calcule a media das avalições do produto a vista(10% do valor Original)
dados_do_lamen['notas'] = pd.to_numeric(dados_do_lamen['notas'],errors = 'coerce')
test1= dados_do_lamen.groupby('marca').agg({'notas':'mean'}).head(50)
data_teste1 = pd.DataFrame(test1).reset_index()
data_teste1['notas'] = data_teste1['notas'] * 1.1
data_teste1
```

```python
#Transforme as palavas das colunas Style para português
#cup = copo , pack = pacote ,tray = bandeja ,bowl = tigela ,box = caixa ,can = lata,bar =barra

dados_do_lamen = dados_do_lamen.replace(['Cup', 'Pack', 'Tray', 'Bowl', 'Box', 'Can', 'Bar'],
                                        ['Copo','Pacote','Bandeja','Tigela','Caixa','Lata','Barra'])
dados_do_lamen
```

```python
#conte quantos pratos (variedades) possuem Noodles
'''
eu não usei o pandas pois pela formula ele gera um Df, e eu queria uma series.
chicago_stations = dados_lamen.df[ dados_lamen.df['variedade'].str.contains('Noodles') ]
chicago_stations.head()
'''
lista_de_Noodles = []
for i in dados_do_lamen.variedade:
    if 'Noodles' in i:
        lista_de_Noodles.append(i)
series_Noodles=pd.Series(lista_de_Noodles)
df_noodles = pd.DataFrame(series_Noodles)
df_noodles
```

```python
#Calcule a quantidade de avaliações por marca
dados_quantidade_marcas = dados_do_lamen.marca.value_counts().to_frame().reset_index().head(10)
dados_quantidade_marcas.rename(columns={'index' : 'marca_unique','marca':'total_marca'},inplace = True)
dados_quantidade_marcas
```

```python
dados_do_lamen.rename(columns={'notas':'notas_puras'},inplace = True)
dados_do_lamen
```

```python
df4 = pd.concat([dados_quantidade_marcas,data_teste1,dados_do_lamen,df_noodles],axis = 1)
# ele coloca um em lado do outro  
df4.rename(columns={0:'marcas_noodles','notas':'medias_notas'},inplace = True)
df4.drop(['top_10'], axis=1,inplace=True)
df4.head(10)
```

```python
df4.to_csv('Relatorio_de_entrega_semana2.csv', index=False)
```
